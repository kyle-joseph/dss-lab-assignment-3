Marben Villaflor: Activity 7: Perform data cleaning and provide a screenshot of both the dirty data (in 5.) and a clean data.

In cleaning the dataset, we remove all the rows that contains "?". From 699 rows, it became 683 rows after cleaning. To remove all the rows with question marks, we loop through all the rows and check each value of the the bare_nuclei column and drop the row using the drop() method of pandas.

```python
for x in dataset.index:
  if dataset.loc[x, "bare_nuclei"] == "?":
    dataset.drop(x, inplace = True)
```
